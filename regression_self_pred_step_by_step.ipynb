```python
def create_dataset_train(dataset, TarY, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back):
        toTrain = dataset[i:(i+look_back), :]
        dataX.append(toTrain)
        toTarY = TarY[i+look_back] ### Y is a single value
        dataY.append(toTarY)
    return np.array(dataX), np.array(dataY)
```

```python
def create_dataset(dataset, TarY, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back):
        toTrain = dataset[i:(i+look_back)]
        dataX.append(toTrain)
        toTarY = TarY[i+look_back] ### Y is a single value
        dataY.append(toTarY)
    return np.array(dataX), np.array(dataY)
```

```python
leadingW_mx_np = np.array(leadingW_mx_df)
MA_ts_np = np.array(df)

col_end = MA_ts_np.shape[1] + 1
row_end = MA_ts_np.shape[0] + 1
look_back =5
look_back_test = look_back-1
look_ahead_test = 1
train_test_ratio = 0.95 ### if it is set to 0.8, it means for training: 0.8, for testing: 0.2
train_row_end = int(MA_ts_np.shape[0] * train_test_ratio + 0.5)
pred_size =  MA_ts_np.shape[0] - train_row_end 
ctrl_val = row_end - 1 ### total prediction time point in testing

### initialize the result matrix
col_pointer = 0
pred_ts_dataMX = pd.DataFrame(MA_ts_np[0:pred_size,col_pointer]).rename(columns = {0:list(df)[col_pointer]})
dataset = df.values
```

```python
### proposed method: leading factors pred target ts
### including self prediction if threre is no leading ts for target ts
### pred t+1 step by step

for j in range(col_pointer,leadingW_mx_np.shape[1]):
        MA_ts_data = pd.DataFrame(MA_ts_np[:,j]).rename(columns = {0:list(df)[j]})               
        pd_tmp = pd.DataFrame(MA_ts_np[0:pred_size,j]) ### given tmp value
        for mv in range(train_row_end, ctrl_val): ### varied 'train_row_end'
            print('mv',mv)
            dataset = MA_ts_data.values
            dataset2 = dataset[0:mv+1,0].astype('float32') # for training
            dataset3 = dataset[mv-look_back_test:mv+look_ahead_test,0].astype('float32') # for testing            
            trainTarY = dataset[0:mv+1,0]
            testTarY = dataset[mv+1:ctrl_val,0]
            
            train = dataset2
            test = dataset3

            trainX, trainY = create_dataset(train,trainTarY, look_back)
            print('trainX:',trainX,'trainY', trainY)
            
            reg = linear_model.LinearRegression()   
            reg.fit(list(trainX[:,:]),list(trainY[:]))
            model_Pred_result= reg.predict(test)
            print('test:',test)
            pd_tmp.iloc[mv-train_row_end] = model_Pred_result
        pred_ts_dataMX[list(df)[j]] = pd_tmp
        pred_ts_dataMX.to_csv('Regression_self_1124.txt', sep='\t', index=0, header = 1, encoding = 'utf8')
```

```python
#è€ƒæ…®ä¸Šå‡ä¸‹é™å¹…åº¦ multi classification(1,0,-1)
TP=0
FP=0
FN=0
TN=0
true_upd_list=[]
pred_upd_list=[]
thresh=0.1
for i in range(0,MA_ts_np.shape[1]): 
    


    t=MA_ts_np[-2,i]#now
    t1_true_origin=MA_ts_np[-1,i]
    t1_pred_origin=pred_ts_dataMX.values[0][i]
    true_upd= (t1_true_origin-t)/t
    pred_upd= (t1_pred_origin-t)/t

    print( 'true:',true_upd,'pred:',pred_upd  )
    if true_upd >=thresh:

        #up

        if pred_upd>=thresh:
            true_upd_list.append(1)
            pred_upd_list.append(1)
            print('1,1')
            #TP+=1
        elif  pred_upd<=-thresh:
            true_upd_list.append(1)
            pred_upd_list.append(-1)
            print('1,-1')
            #FN+=1
        else:
            true_upd_list.append(1)
            pred_upd_list.append(0)
            print('1,0')

    elif true_upd<=-thresh:
        #down
        if pred_upd>=thresh:
            true_upd_list.append(-1)
            pred_upd_list.append(1)
            print('-1,1')
            #FP+=1
        elif  pred_upd<=-thresh:
            true_upd_list.append(-1)
            pred_upd_list.append(-1)
            print('-1,-1')
            #TN+=1
        else :
            true_upd_list.append(-1)
            pred_upd_list.append(0)
            print('-1,0')
    else:

        if pred_upd>=thresh:
            true_upd_list.append(0)
            pred_upd_list.append(1)
            print('0,1')
            #FP+=1
        elif  pred_upd<=-thresh:
            true_upd_list.append(0)
            pred_upd_list.append(-1)
            print('0,-1')
            #TN+=1
        else :
            true_upd_list.append(0)
            pred_upd_list.append(0)
            print('0,0')
```

```python
from sklearn import metrics
print(metrics.precision_score(true_upd_list, pred_upd_list, average='macro'))
print(metrics.recall_score(true_upd_list, pred_upd_list, average='macro'))
print(metrics.f1_score(true_upd_list, pred_upd_list, average='micro'))
```
